## Import image processing components

```python
from skimage.io import imread,imshow
from skimage.transform import resize
from skimage.color import rgb2gray   
```

## Setting the Working Directory 

```python
cars=os.listdir('Imagesclassify/Images/Cars/')
cars
```

```python
photos = os.listdir('Imagesclassify/Images/Photos/')
naruto = os.listdir('Imagesclassify/Images/Naruto/')
```

## Reading images

```python
limit = 30
cars_images = [None]*limit
j = 0
for i in cars:
    if j<limit:
        cars_images[j] = imread("Imagesclassify/Images/Cars/"+i)
        j = j+1
    else:
        break
```

```python
imshow(cars_images[24])
```

```python
limit = 30
naruto_images = [None]*limit
j = 0
for i in naruto:
    if j<limit:
        naruto_images[j] = imread("Imagesclassify/Images/Naruto/"+i)
        j = j+1
    else:
        break
```

```python
imshow(naruto_images[29])
```

```python
limit = 30
photos_images = [None]*limit
j = 0
for i in photos:
    if j<limit:
        photos_images[j] = imread("Imagesclassify/Images/Photos/"+i)
        j = j+1
    else:
        break
```

```python
imshow(photos_images[23])
```

```python
photos_images[6].shape
```

```python
photos_images[6]
```

## Convert the image colour images to greyscale

```python
cars_gray=[None]*limit
j=0
for i in cars:
    if(j<limit):
        cars_gray[j]=rgb2gray(cars_images[j])
        j+=1
    else:
        break
```

```python
naruto_gray=[None]*limit
j=0
for i in naruto:
    if(j<limit):
        naruto_gray[j]=rgb2gray(naruto_images[j])
        j+=1
    else:
        break
```

```python
photos_gray=[None]*limit
j=0
for i in photos:
    if(j<limit):
        photos_gray[j]=rgb2gray(photos_images[j])
        j+=1
    else:
        break
```

```python
cars_gray[1]
```

```python
cars_gray[1].shape
```

```python
photos_gray[4].shape
```

```python
naruto_gray[12].shape
```

## Matrix Resizing

```python
for j in range(30):
    cr=cars_gray[j]
    cars_gray[j]=resize(cr,(512,512))
```

```python
for j in range(30):
    nr=naruto_gray[j]
    naruto_gray[j]=resize(nr,(512,512))
```

```python
for j in range(30):
    ph=photos_gray[j]
    photos_gray[j]=resize(ph,(512,512))
```

```python
cars_gray[3].shape
```

```python
naruto_gray[2].shape
```

```python
photos_gray[4].shape
```

## Image matrix to vector conversion (512  x 512 to 262144 x 1)

### Step 1 : check for the length of the images

```python
len_images_cars = len(cars_gray)
```

```python
len_images_cars
```

### Step 2 :check for the image size

```python
image_size_cars = cars_gray[4].shape
```

```python
image_size_cars 
```

### Step 3: Create a variable for the product of image size

```python
flatten_size = image_size_cars[0]*image_size_cars[1]
```

```python
flatten_size
```

### Step 4 : Flatten the images

```python
for i in range(30):
    cars_gray[i] = np.ndarray.flatten(cars_gray[i]).reshape(flatten_size,1)
```

```python
cars_gray[4].shape
```

### Step 5: Stacking the images

```python
cars_gray
```

```python
cars_gray = np.dstack(cars_gray)
```

```python
cars_gray.shape
```

### Step 6 :Rolling the axis of the images

```python
cars_gray = np.rollaxis(cars_gray, axis = 2, start=0)
```

```python
cars_gray.shape
```

### Step 7 : reshaping to remove the colour channel

```python
cars_gray = cars_gray.reshape(30,262144)
```

```python
cars_gray.shape
```

### Step 8:upload the data into a dataframe

```python
cars_data = pd.DataFrame(cars_gray)
```

```python
cars_data
```

### Step 9: label the data

```python
cars_data['label']='cars'
```

```python
cars_data
```
